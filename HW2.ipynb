```python
%%latex
Prove
\begin{equation}
\\
\left\lVert A^{-1} \right\rVert = \max_{u\neq 0} \frac{\left\lVert u \right\rVert}{\left\lVert Au \right\rVert}
\\
\end{equation}

We know from the definition of a norm that

\begin{equation}
\\
\left\lVert A \right\rVert = \max_{v\neq 0} \frac{\left\lVert Av \right\rVert}{\left\lVert v \right\rVert}
\\
\end{equation}

Therefore

\begin{equation}
\\
\left\lVert A^{-1} \right\rVert = \max_{v\neq 0} \frac{\left\lVert A^{-1}v \right\rVert}{\left\lVert v \right\rVert}
\\
\end{equation}

If, however, we define

\begin{equation}
\\
u = Av
\\
\end{equation}

Then we end up with

\begin{equation}
\\
\left\lVert A^{-1} \right\rVert = \max_{u\neq 0} \frac{\left\lVert A^{-1}Au \right\rVert}{\left\lVert Au \right\rVert}
\\
\end{equation}

Which, when we simplify, gives

\begin{equation}
\\
\left\lVert A^{-1} \right\rVert = \max_{u\neq 0} \frac{\left\lVert u \right\rVert}{\left\lVert Au \right\rVert}
\\
\end{equation}

However, finding the maximum of a non-negative function over some space is equivalent to finding the minimum of the reciprocal of that same function over the same space. Therefore, since

\begin{equation}
\\
\max_{x} f\left(x\right) = \frac{1}{\min_x f^{-1}\left(x\right)}
\\
\end{equation}

We have proven

\begin{equation}
\\
\left\lVert A^{-1} \right\rVert = \left(\min_{u\neq 0} \frac{\left\lVert Au \right\rVert}{\left\lVert u \right\rVert}\right)
\\
\end{equation}
```

```python
%%latex

Prove
\begin{equation}
\\
\left\lVert Au \right\rVert \leq \left\lVert A \right\rVert \left\lVert u \right\rVert
\\
\end{equation}

Once more, the definition is as follows


\begin{equation}
\\
\left\lVert A \right\rVert = \max_{u\neq 0} \frac{\left\lVert Au \right\rVert}{\left\lVert u \right\rVert}
\\
\end{equation}

Reasonably, therefore, since $\left\lVert A\right\rVert$ is the maximum possible value of
$\frac{\lVert Au\rVert}{\lVert u\rVert}$, the following inequality must be true

\begin{equation}
\\
\left\lVert A \right\rVert \geq \frac{\left\lVert Au \right\rVert}{\left\lVert u \right\rVert}
\\
\end{equation}

We can multiply both sides by $\lVert u\rVert$, giving

\begin{equation}
\\
\lVert A \rVert \lVert u \rVert \geq \frac{\lVert Au \rVert}{\lVert u \rVert}\lVert u \rVert
\\
\end{equation}

Which can be simplied down to

\begin{equation}
\\
\lVert A \rVert \lVert u \rVert \geq \lVert Au \rVert
\\
\end{equation}
```

```python
%%latex

Prove
\begin{equation}
\\
\lVert bA \rVert = \left|b\right| \lVert A\rVert
\\
\end{equation}

By the definition of the operator norm

\begin{equation}
\\
\left\lVert bA \right\rVert = \max_{u\neq 0} \frac{\lVert bAu \rVert}{\lVert u \rVert}
\\
\end{equation}

For a vector, we know $\lVert cu\rVert = \left|c\right| \lVert u\rVert$, where $c$ is a scalar. Therefore

\begin{equation}
\\
\left\lVert bA \right\rVert = \max_{u\neq 0} \frac{\left|b\right|\lVert Au \rVert}{\lVert u \rVert}
\\
\end{equation}

Since $\left|b\right|$ is invariant with $u$, this is equivalent to
\begin{equation}
\\
\left\lVert bA \right\rVert = \left|b\right|\max_{u\neq 0} \frac{\lVert Au \rVert}{\lVert u \rVert}
\\
\end{equation}

Therefore
\begin{equation}
\\
\lVert bA \rVert = \left|b\right| \lVert A\rVert
\\
\end{equation}
```

```python
%%latex

Prove
\begin{equation}
\\
\lVert A + B \rVert \leq \lVert A\rVert + \lVert B \rVert
\\
\end{equation}

By the definition of the operator norm

\begin{equation}
\\
\left\lVert A+B \right\rVert = \max_{u\neq 0} \frac{\left\lVert \left(A+B\right)u \right\rVert}{\left\lVert u \right\rVert}
\\
\end{equation}

Distribution gives

\begin{equation}
\\
\left\lVert A+B \right\rVert = \max_{u\neq 0} \frac{\left\lVert Au+Bu \right\rVert}{\left\lVert u \right\rVert}
\\
\end{equation}

Minkowski's inequality tells us $\lVert u + v \rVert \leq \lVert u \rVert + \lVert v\rVert$ when $u$ and $v$ are vectors.
Since $Au$ and $Bu$ are vectors, Minkowski's inequality holds true here, thus:

\begin{equation}
\\
\left\lVert A+B \right\rVert \leq \max_{u\neq 0} \frac{\lVert Au\rVert +\lVert Bu \rVert}{\left\lVert u \right\rVert}
\\
\end{equation}

\begin{equation}
\\
\left\lVert A+B \right\rVert \leq \max_{u\neq 0}\left( \frac{\lVert Au\rVert}{\lVert u\rVert} +\frac{\lVert Bu \rVert}{\left\lVert u \right\rVert}\right)
\\
\end{equation}

This yields

\begin{equation}
\\
\lVert A + B \rVert \leq \lVert A\rVert + \lVert B \rVert
\\
\end{equation}
```

```python
%%latex
Prove

\begin{equation}
\\
\lVert AB \rVert \leq \lVert A\rVert \lVert B \rVert
\\
\end{equation}

By definition

\begin{equation}
\\
\left\lVert AB \right\rVert = \max_{v\neq 0} \frac{\left\lVert ABv \right\rVert}{\lVert v\rVert}
\\
\end{equation}

\begin{equation}
\\
\left\lVert AB \right\rVert = \max_{v\neq 0} \frac{\left\lVert ABv \right\rVert}{\lVert v\rVert}\frac{\lVert Bv\rVert}{\lVert Bv\rVert}
\\
\end{equation}

\begin{equation}
\\
\left\lVert AB \right\rVert = \max_{v\neq 0} \frac{\left\lVert ABv \right\rVert}{\lVert Bv\rVert}\frac{\lVert Bv\rVert}{\lVert v\rVert}
\\
\end{equation}

Defining $Bv = u$

\begin{equation}
\\
\left\lVert AB \right\rVert = \max_{v\neq 0} \frac{\left\lVert Au \right\rVert}{\lVert u\rVert}\frac{\lVert Bv\rVert}{\lVert v\rVert}
\\
\end{equation}

Based on how multiplication works, the following inequality must be true

\begin{equation}
\\
\max_{v\neq 0} \frac{\left\lVert Au \right\rVert}{\lVert u\rVert}\frac{\lVert Bv\rVert}{\lVert v\rVert} \leq \max_{u\neq 0} \frac{\left\lVert Au \right\rVert}{\lVert u\rVert}\max_{v\neq0}\frac{\lVert Bv\rVert}{\lVert v\rVert}
\\
\end{equation}

Therefore

\begin{equation}
\\
\left\lVert AB \right\rVert \leq \max_{u\neq 0} \frac{\left\lVert Au \right\rVert}{\lVert u\rVert}\max_{v\neq0}\frac{\lVert Bv\rVert}{\lVert v\rVert}
\\
\end{equation}

Which, when we substitute using definitions, gives 

\begin{equation}
\\
\lVert AB \rVert \leq \lVert A\rVert \lVert B \rVert
\\
\end{equation}
```

```python
%%latex
\begin{equation}
\\
A = \begin{bmatrix}
    1 & \rho \\
    \rho & 1
\end{bmatrix}
\\
\end{equation}

The Cholesky decomposition requires figuring out the $LDL^T$ decomposition, which gives the following

\begin{equation}
\\
\begin{bmatrix}
    1 & \rho \\
    \rho & 1
\end{bmatrix} = 
\begin{bmatrix}
    1 & 0 \\
    L & 1 
\end{bmatrix}
\begin{bmatrix}
    D_1 & 0 \\
    0 & D_2
\end{bmatrix}
\begin{bmatrix}
    1 & L \\
    0 & 1 
\end{bmatrix}
\\
\end{equation}

Simplifying, we get

\begin{equation}
\\
\begin{bmatrix}
    1 & \rho \\
    \rho & 1
\end{bmatrix} = 
\begin{bmatrix}
    D_1 & D_1L \\
    D_1L & D_1L^2 + D_2 
\end{bmatrix}
\\
\end{equation}

From this, it follows that

\begin{equation}
D_1 = 1 \\
L = \rho \\
D_2 = 1 - \rho^2 \\
\end{equation}

The Cholesky decomposition is $L\sqrt{D}$, which means we have

\begin{equation}
\\
L\sqrt{D} = 
\begin{bmatrix}
    1 & 0 \\
    \rho & 1
\end{bmatrix}
\begin{bmatrix}
    1 & 0 \\
    0 & \sqrt{1 - \rho^2}
\end{bmatrix}
\\
\end{equation}

\begin{equation}
\\
L\sqrt{D} = 
\begin{bmatrix}
    1 & 0 \\
    \rho & \sqrt{1 - \rho^2}
\end{bmatrix}
\\
\end{equation}
```

```python
### Part 2

import numpy as np

def cholesky_decomp(mat):
    for r1, r2 in zip(mat, mat.transpose()):
        assert len(r1) == len(r2), "Input matrix is not symmetric"
        assert (r1 == r2).all(), "Input matrix is not symmetric"

    assert all(x > 0 for x in np.linalg.eig(mat)[0]), "Input matrix is not positive definite"

    output = mat.copy()

    def recurse(m):
        if len(m) == 1:
            m[0,0] = np.sqrt(m[0,0])
            return

        p = m[0,0] = np.sqrt(m[0,0])
        L = m[1:, 0] = m[1:, 0] / p
        m[0, 1:] = 0
        m[1:, 1:] = m[1:, 1:] - L*L.transpose()
        recurse(m[1:, 1:])

    recurse(output)
    return output

mat = np.mat([[4,12,-16],[12,37,-43],[-16,-43,98]])
cholesky_decomp(mat)
```

```python
### Part 3
import random
import numpy as np
import matplotlib.pyplot as plt

def generate_matrix(size):
    d = np.mat([[0 if i!=j else size for j in range(size)] for i in range(size)])
    A = np.random.rand(size, size)
    return 0.5* (A+A.transpose()) + d


def cholesky_decomp(mat):
    for r1, r2 in zip(mat, mat.transpose()):
        assert len(r1) == len(r2), "Input matrix is not symmetric"
        assert ((r1-r2)/r2 < 0.01).all(), "Input matrix is not symmetric"

    assert all(x > 0 for x in np.linalg.eig(mat)[0]), "Input matrix is not positive definite"

    output = mat.copy()

    def recurse(m):
        if len(m) == 1:
            m[0,0] = np.sqrt(m[0,0])
            return

        p = m[0,0] = np.sqrt(m[0,0])
        L = m[1:, 0] = m[1:, 0] / p
        m[0, 1:] = 0
        m[1:, 1:] = m[1:, 1:] - L*L.transpose()
        recurse(m[1:, 1:])

    recurse(output)
    return output

sizes = []
np_times = []
hrk_times = []

for i in range(1, 8):
    print(i)
    size = 2**i
    sizes.append(size)
    m = generate_matrix(size)

    x = %timeit -q -n 100 -o np.linalg.cholesky(m)
    np_times.append(x.average)
    y = %timeit -q -n 100 -o cholesky_decomp(m)
    hrk_times.append(y.average)

x_axis = np.array(sizes)
numpy_res = np.array(np_times)
hrk_res = np.array(hrk_times)
plt.plot(x_axis, numpy_res, label='NUMPY')
plt.plot(x_axis, hrk_res, label='STUDENT')
plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.xlabel('n')
plt.ylabel('Decomposition Time (s)')
plt.figure()
plt.show()
```

```python
import numpy as np
import pandas as pd

dataurl = "https://raw.githubusercontent.com/yadongli/nyumath2048/master/data/djiahist.csv"
histprice = pd.read_csv(dataurl, index_col=[0]).sort_index()

djiaurl = "https://raw.githubusercontent.com/yadongli/nyumath2048/master/data/djia.csv"
djia = pd.read_csv(djiaurl, index_col=[0])
```

```python
### Part 1
def find_covariance_matrix(histprice):
    n = len(histprice)
    A = histprice - np.ones([n,n]).dot(histprice)/n
    return A.transpose().dot(A)/n

def find_correlation_matrix(histprice):
    cov = find_covariance_matrix(histprice)
    diag = cov.copy()
    for i in range(len(diag)):
        for j in range(len(diag)):
            if i != j:
                diag[i][j] = 0
            else:
                diag[i][j] = np.sqrt(1/diag[i][j])

    return diag.dot(cov).dot(diag)


daily_returns = np.log(histprice).diff().values[1:]

annualized_daily_returns = daily_returns * 252;

cov = find_covariance_matrix(annualized_daily_returns)
corr = find_correlation_matrix(annualized_daily_returns)
print(cov)
print(corr)
```

```python
### Part 2
def calculate_excess_returns(risk_free_rate, annualized_daily_returns):
    n_obs = len(annualized_daily_returns)
    avg_returns = np.ones([1, n_obs]).dot(annualized_daily_returns) / n_obs
    excess = avg_returns - np.log(1+risk_free_rate)
    return pd.DataFrame(excess, columns=histprice.columns)
    

risk_free_rate = 0.03
excess_returns = calculate_excess_returns(risk_free_rate, annualized_daily_returns)
excess_returns
```

```python
### Part 3
cov_inverse = np.linalg.inv(cov)
er = excess_returns.values
w = cov_inverse.dot(er.T)/er.dot(cov_inverse).dot(er.T)
w /= np.sum(w)
pd.DataFrame(w.T, columns=histprice.columns)
```

```python
%%latex
The portfolio looks pretty reasonable, all things considered. There are several individual positions that are
larger than the total value of the portfolio, which is risky, but the differences in excess annualized returns
are also quite large, so one would expect a fairly large difference.
```

```python
### Part 4
djia_w = djia.Weights.values
vb = djia_w.dot(cov).dot(djia_w)
implied_returns = cov.dot(djia_w) / vb

percent_returns = pd.DataFrame([implied_returns],columns=histprice.columns)
percent_returns
log_implied_returns = np.log(implied_returns+1)
pd.DataFrame([log_implied_returns], columns=histprice.columns)

```

```python
### Part 5
condition = np.linalg.norm(cov)*np.linalg.norm(cov_inverse)
print(condition)
```

```python
%%latex
The condition number of the covariance matrix is 217. This implies a good deal of sensitivity to perturbation. Relying
on this model for maintaining a portfolio may be unfeasible.
```

```python
### Part 6
evens = histprice[::2]
evens_daily_returns = np.log(evens).diff().values[1:]
evens_average_daily_returns = np.ones([1, len(evens_daily_returns)]).dot(evens_daily_returns*126) / len(evens_daily_returns)

odds = histprice[1::2]
odds_daily_returns = np.log(odds).diff().values[1:]
odds_average_daily_returns = np.ones([1, len(odds_daily_returns)]).dot(odds_daily_returns*126) / len(odds_daily_returns)

pd.DataFrame(100*(evens_average_daily_returns - odds_average_daily_returns) / odds_average_daily_returns, columns=histprice.columns)
```

```python
evens_vol = np.std(evens_daily_returns)*np.sqrt(126)
odds_vol = np.std(odds_daily_returns)*np.sqrt(126)

(evens_vol - odds_vol) / odds_vol
```

```python
%%latex
The volatility has a 2% error. Most of the returns have errors on approximately the same order of magnitude.
However, some of the returns vary by more than 100%, which tells us that historical returns are actually a 
poor estimator. This makes some sense as we know that, in a random walk such as the stock market, past performance
is not a reliable predictor of future performance
```
